```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
```

```python
# 그래프 그리기 연습
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('First_plot')
plt.figure(figsize = (15,15))
fig.savefig('first_plot.png')
```

```python
#np.arange는 실수도 됨 사칙연산도 됨
fig , ax = plt.subplots()
x = np.arange(10)
#실선 solid
ax.plot(x,x,linestyle = '-')
#대시선 dashed
ax.plot(x,x+2,linestyle = '--')
#대시 점 선 dashdot
ax.plot(x,x+4,linestyle = '-.')
#점선 dotted
ax.plot(x,x+6,linestyle = ':')
```

```python
x = np.arange(10)
fig, ax = plt.subplots()
# RGB, CMYK
ax.plot(x, x, color = 'r')
ax.plot(x, x+2, color = 'C')
# grayscale 0 검정색, 1이면 흰색
ax.plot(x, x+4, color = '0.5')
# 색상에 대한 16진수 코드
ax.plot(x, x+6, color = '#0f1629')
```

```python
fig, ax = plt.subplots()
x = np.arange(10)

ax.plot(x, x, marker = '.')
ax.plot(x, x+2, marker = 'o')
ax.plot(x, x+4, marker = '^')
ax.plot(x, x+6, marker = 's')
ax.plot(x, x+8, marker = '*')
```

기호 의미
. 점
, 픽셀
o 원
s 사각형
v, <, ^, > 삼각형
1, 2, 3, 4 삼각선
p 오각형
H, h 육각형

```python
fig, ax = plt.subplots()
x = np.linspace(0, 10 ,1000) #0부터 10까지  1000개의 데이터 가져옴

ax.plot(x, np.sin(x))

ax.set_xlim(-2,12)
ax.set_ylim(-1.5,1.5)
```

```python
fig, ax = plt.subplots()
x = np.arange(10)

ax.plot(x, x, label='y=x')
ax.plot(x, x**2, label='y=x^2')

ax.set_xlabel('x')
ax.set_ylabel('y', rotation = 0)

ax.legend(loc = 'upper left', shadow = True, fancybox = True, borderpad = 2) #upper, lower, left, right, center
```

```python
fig, ax = plt.subplots()
x = np.arange(10)

ax.plot(x, x, label='y=x', color = 'b', marker = 'o', linestyle = ':')
ax.plot(x, x**2, label = 'y=x^2', color = 'r', marker = '^', linestyle = '--')

ax.set_xlabel('x')
ax.set_ylabel('y',rotation=0)

ax.legend(loc = 'upper left',borderpad=2,shadow = True,fancybox = True)

fig.savefig('test.png')
```

```python
fig, ax = plt.subplots()
x = np.random.randn(50) #randn은 평균이 0이 되는 랜덤값
y = np.random.randn(50)
colors = np.random.randint(0, 100, 50) #시작부터 n-1까지의 수 안에서 랜덤 값 
sizes = 500 * np.pi * np.random.rand(50) **2 #rand는 0~1사이의 실수값
ax.scatter(
    x,y,c = colors, s = sizes, alpha = 0.3
)
```

```python
x = np.random.rand(5)
y = np.random.rand(5)
z = np.random.rand(5)

data  = [x,y,z]
fig, ax = plt.subplots()

x_ax = np.arange(5) # [0,1,2]
for i in range(len(data)): 
  ax.bar(x_ax,data[i],bottom = np.sum(data[:i],axis=0)) # i 번째까지 더해서 그래프 밑에서부터 쌓아줌
ax.set_xticks(x_ax)
# ax.set_xticklabels(['A','B','C',''])
```
