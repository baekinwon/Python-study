```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
import matplotlib.pyplot as plt
```

```python
# 붓꽃의 꽃잎(petal)과 꽃받침(sepal)의 폭과 길이를 센티미터 단위로 측정
# 전문 식물학자가 setosa, versicolor, virginica 종으로 분류
```

```python
#사이킷런에서 붓꽃 품종 데이터 셋트 가져옴
dataset = datasets.load_iris()
dataset
```

```python
#입력 데이터와 타깃을 준비합니다.
X , y = dataset['data'],dataset['target']
```

```python
# 데이터세트를 학습 세트와 테스트 세트로 분할합니다.
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0)
```

```python
# 결정 트리 모델 객체 만듭니다.
model = DecisionTreeClassifier()
```

```python
#결정 트리 모델에 학습 세트의 입력 데이터와 타깃을 넣고 학습시킵니다.
model.fit(X_train,y_train)
```

```python
#학습세트에서 모델 정확도와 테스트 세트에서의 모델 정확도를 계산합니다.
train_score = model.score(X_train,y_train)
test_score = model.score(X_test,y_test)

#학습 세트에서의 정확도와 테스트 세트에서의 정확도를 출력합니다.
print('학습 세트 정확도 : {:.3f}'.format(train_score))
print('테스트 세트 정확도 : {:.3f}'.format(test_score))
```

```python
# 학습된 결정 트리를 시각화합니다.
plot_tree(model,filled=True)
plt.show()
```

```python
# 과적합 문제 해결
model = DecisionTreeClassifier(max_depth=3)
```

```python
model.fit(X_train,y_train)

#학습세트에서 모델 정확도와 테스트 세트에서의 모델 정확도를 계산합니다.
train_score = model.score(X_train,y_train)
test_score = model.score(X_test,y_test)

#학습 세트에서의 정확도와 테스트 세트에서의 정확도를 출력합니다.
print('학습 세트 정확도 : {:.3f}'.format(train_score))
print('테스트 세트 정확도 : {:.3f}'.format(test_score))
```

```python
# 학습된 결정 트리를 시각화합니다.
plot_tree(model,filled=True)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train,y_train)
train_score = lr.score(X_train,y_train)
test_score = lr.score(X_test,y_test)
print('학습 세트 정확도 : {:.3f}'.format(train_score))
print('테스트 세트 정확도 : {:.3f}'.format(test_score))
```
