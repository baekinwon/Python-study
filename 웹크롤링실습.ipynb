```python
from selenium import webdriver

# colab에서 selenium을 돌리기 위한 옵션들
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
```

```python
#request 불러오기

import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from selenium import webdriver  # 셀레니움을 활성화
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python

#headers에 'user agent' 값 넣기
#구글에서 my user agent 검색해서 가져오기

headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.82 Safari/537.36'}

#url 주소 가져오기
#첫페이지 정보 가져오기

url = 'https://www.yanolja.com/leisure/list?mediumcat=10120013'

#request.get()으로 요청하기

r = requests.get(url, headers = headers)

#200이 나오면 연결성공
r.status_code 
```

```python
driver = webdriver.Chrome("chromedriver", options=chrome_options)
driver.get(url)
```

```python
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
```

```python
driver.page_source
```

```python
#페이지의 정보를 출력
soup = BeautifulSoup(r.text,'html.parser')
soup
```

```python
site_info = soup.find_all('a',class_='LeisureListItem_container__12YuE')
```

```python
print(site_info)
```

```python
print(len(site_info))
```

```python
name = []
price = []
```

```python
for info in site_info:
  site_name = info.find_all('p','LeisureListItem_title__U-d8s')
  name.append(site_name[0].text)
  site_price = info.find_all('span',class_='LeisurePriceDiscount_amount__19qBa')
  price.append(site_price[0].text)
```

```python
df = pd.DataFrame(
    {'name' : name,
     'price' : price}
)
df
```

```python
df.dtypes
```

```python
df['price'] = df['price'].str.replace(pat=r',',repl=r'',regex=True)
```

```python
df['price']= df['price'].astype('int')
```

```python
df = df.sort_values(by = 'price')
```

```python
df.to_csv('30713_백인원.csv',index=False,encoding="utf-8-sig")
```
