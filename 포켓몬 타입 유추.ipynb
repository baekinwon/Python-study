```python
import pandas as pd
pokemon = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/머신러닝/자료/pokemon.csv')
pokemon.head() #포켓몬 csv파일 불러옴
```

```python
pokemon.isnull().sum()
```

```python
poke_grass_type = pokemon[(pokemon['main_type']=='Grass') | (pokemon['secondary_type']=='Grass')]
poke_grass_type.head(5) #풀타입 포켓몬 가져오기
```

```python
grass_height = pd.DataFrame(poke_grass_type['height'])
grass_height = pd.DataFrame(grass_height['height'].str.strip("m"))
grass_height = pd.DataFrame(grass_height['height'].astype(float)).values.tolist()
grass_height#포켓몬의 키를 float형으로 변환해 리스트로 가져옴
```

```python
grass_weight = pd.DataFrame(poke_grass_type['weight'])
grass_weight = pd.DataFrame(grass_weight['weight'].str.strip("kg"))
grass_weight = pd.DataFrame(grass_weight['weight'].astype(float)).values.tolist()
grass_weight#포켓몬의 무게를 float형으로 변환해 리스트로 가져옴
```

```python
poke_rock_type = pokemon[(pokemon['main_type']=='Rock') | (pokemon['secondary_type']=='Rock')]
poke_rock_type.head(5) #바위 타입 포켓몬 가져오기
```

```python
rock_height = pd.DataFrame(poke_rock_type['height'])
rock_height = pd.DataFrame(rock_height['height'].str.strip("m"))
rock_height = pd.DataFrame(rock_height['height'].astype(float)).values.tolist()
rock_height#포켓몬의 키를 float형으로 변환해 리스트로 가져옴
```

```python
rock_weight = pd.DataFrame(poke_rock_type['weight']) #무게를 데이터 프레임 형식으로 불러옴
rock_weight = pd.DataFrame(rock_weight['weight'].str.strip("kg")) #kg이라 붙은 문자열을 지움
rock_weight = pd.DataFrame(rock_weight['weight'].astype(float)).values.tolist()
rock_weight #포켓몬의 무게를 float형으로 변환해 리스트로 가져옴
```

```python
all_height = grass_height+ rock_height
all_height #두 타입의 모든 키
```

```python
all_height = sum(all_height,[])
all_height #두 타입의 모든 키를 1차원리스트로 변환
```

```python
all_weight = grass_weight+ rock_weight
all_weight   #두 타입의 모든 무게
```

```python
all_weight = sum(all_weight,[])
all_weight #두타입의 모든 무게를 1차원리스트로 변환
```

```python
import numpy as np
poke_data = np.column_stack((all_height,all_weight))
print(poke_data[:5])   #키와 몸무게로 훈련할 데이터 2차원으로 다시 배열
poke_data.shape
```

```python
poke_target = np.concatenate((np.ones(91),np.zeros(57))) #풀타입 91개 바위타입 57개
print(poke_target)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_input, test_input, train_target, test_target = train_test_split(
    poke_data,poke_target,random_state =64)
print(train_input.shape,test_input.shape)
print(train_target.shape,test_target.shape)
```

```python
from pandas.core.common import random_state
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier()
kn.fit(train_input,train_target)
kn.score(test_input,test_target) 
```

```python
if kn.predict([[0.9,6.4]])==1: #치코리타
  print('풀타입')
else:
  print("바위타입")

if kn.predict([[2,202]])==1: #마기라스
  print('풀타입')
else:
  print("바위타입")
```

KNN 알고리즘 써보니 정확도가 매우 낮은 문제 발생
따라서 여러 알고리즘 사용해봄

```python
from sklearn.metrics import accuracy_score # 정확도 함수
from sklearn.linear_model import LogisticRegression #로지스틱회귀가 가장 정확도가 높음
lr = LogisticRegression()
lr.fit(train_input,train_target)
predict1 = lr.predict(test_input)
print(accuracy_score(test_target,predict1))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold 


params = { 'n_estimators' : [10, 20, 100],
           'max_depth' : [3, 6, 8, 10, 12],
           'min_samples_leaf' : [8, 12, 18],
           'min_samples_split' : [8, 16, 20]
            }

# RandomForestClassifier 객체 생성 후 GridSearchCV 수행
clf = RandomForestClassifier(random_state = 0, n_jobs = -1)
grid_cv = GridSearchCV(clf,param_grid = params, cv = 3, n_jobs = -1)
grid_cv.fit(train_input, train_target)

print('최적 하이퍼 파라미터: ', grid_cv.best_params_)
print('최고 예측 정확도: {:.4f}'.format(grid_cv.best_score_))
```

```python
clf_s = RandomForestClassifier(n_estimators=100, max_depth=3,random_state=16,min_samples_leaf = 8,min_samples_split = 20) 
clf_s.fit(train_input,train_target)
predict1 = clf_s.predict(test_input)
print(accuracy_score(test_target,predict1))
```

```python
from xgboost import XGBClassifier
params = { 
           'eta' : [0.1 , 0.2 , 0.3],
           'n_setimators' : [100, 200, 300, 400, 500],
           'max_depth' : [3, 6, 8, 10, 12],
           'min_child_weight' : [5, 10, 15],
           'learning_rate' : [0.1, 0.2, 0.3]
            }

# RandomForestClassifier 객체 생성 후 GridSearchCV 수행
xgb = XGBClassifier(random_state = 0, n_jobs = -1)
grid_cv = GridSearchCV(xgb, param_grid = params, cv = 3, n_jobs = -1)
grid_cv.fit(train_input, train_target)

print('최적 하이퍼 파라미터: ', grid_cv.best_params_)
print('최고 예측 정확도: {:.4f}'.format(grid_cv.best_score_))
```

```python

params = {
    'eta': 0.1,
    'max_depth':3, #깊이를 3으로 지정
    'learning_rate':0.1,#가중치 0.1로 지정 learning_rate는 값이 높을수록 과적합 쉬움
    'n_setimators':100, # learning_rate가 적으면 적을수록 n_setimators값이 높아야 과적합 방지
    'min_child_weight':5, #과적합 방지를 위해 높은값을 사용한다.
}
model = XGBClassifier(params=params) 
model.fit(train_input,train_target)
predict = model.predict(test_input)
print(accuracy_score(test_target,predict))
```
