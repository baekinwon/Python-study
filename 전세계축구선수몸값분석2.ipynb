```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.82 Safari/537.36'}
number = []
name = []
position = []
age = []
nation = []
team = []
value = []
for i in range(1,3):
  url = 'https://www.transfermarkt.com/spieler-statistik/wertvollstespieler/marktwertetop?page='+str(i)

  r = requests.get(url, headers = headers)

  soup = BeautifulSoup(r.text,'html.parser')

  player_info = soup.find_all('tr',class_ = ['odd','even']) 

  for info in player_info:
    players = info.find_all('td')
    number.append(players[0].text)
    name.append(players[3].text)
    position.append(players[4].text)
    age.append(players[5].text)
    nation.append(players[6].img['alt'])
    team.append(players[7].img['alt'])
    value.append(players[8].text)

  time.sleep(1)

tf= pd.DataFrame(
    { 'number' : number,
     'name' : name,
     'position' : position,
     'age' : age,
     'nation' : nation,
     'team' : team,
     'value' : value}
)

tf
#tf를 csv파일로 저장
tf.to_csv('transfermakt50.csv', index = False)
```
